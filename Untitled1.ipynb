```python
''' Importación de librerías '''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
''' Carga de datos '''

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

data = requests.get(url).json()

df = pd.json_normalize(data)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.rename(columns={
    "customer.gender": "gender",
    "customer.SeniorCitizen": "SeniorCitizen",
    "customer.Partner": "Partner",
    "customer.Dependents": "Dependents",
    "account.tenure": "tenure",
    "account.Contract": "Contract",
    "account.PaperlessBilling": "PaperlessBilling",
    "account.PaymentMethod": "PaymentMethod",
    "account.Charges.Monthly": "MonthlyCharges",
    "account.Charges.Total": "TotalCharges"
})
```

```python
''' inconsistencias '''

df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["Churn"].unique()
```

```python
df["Churn"] = df["Churn"].map({"Yes":1, "No":0})
```

```python
df["Cuentas_Diarias"] = df["MonthlyCharges"] / 30
```

```python
df.describe()
```

```python
sns.countplot(x="Churn", data=df)
plt.title("Distribución de Evasión de Clientes")
plt.show()
```

```python
sns.countplot(x="Contract", hue="Churn", data=df)
plt.title("Evasión según tipo de contrato")
plt.show()
```

```python
sns.countplot(x="PaymentMethod", hue="Churn", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Churn vs Monthly Charges")
plt.show()
```

```python
sns.boxplot(x="Churn", y="customer.tenure", data=df)
plt.title("Churn vs Tiempo de contrato")
plt.show()
```

# informe final

El presente análisis tiene como objetivo estudiar el fenómeno de evasión de clientes en la empresa. La evasión de clientes ocurre cuando un usuario decide cancelar los servicios de la compañía, lo que representa una pérdida directa de ingresos y puede afectar la estabilidad financiera de la empresa.

# Limpieza y Tratamiento de Datos

Para realizar el análisis fue necesario seguir un proceso de preparación de los datos que incluyó varias etapas importantes.

Primero, los datos fueron extraídos desde una API en formato JSON alojada en el repositorio de GitHub del desafío. Posteriormente, se utilizó la librería Pandas para convertir la estructura JSON en un DataFrame, lo que permitió manipular y analizar los datos de manera más eficiente.

Debido a que el archivo JSON contenía estructuras anidadas, fue necesario normalizar los datos y renombrar algunas columnas para facilitar su interpretación. Además, se realizaron verificaciones para identificar valores faltantes, posibles duplicados y errores de formato.

# Análisis Exploratorio de Datos

En la fase de análisis exploratorio se utilizaron diversas visualizaciones para comprender mejor el comportamiento de los clientes y detectar patrones relacionados con la evasión.

El análisis de la distribución de la variable Churn permitió observar la proporción de clientes que permanecen en la empresa frente a aquellos que cancelan el servicio.

Posteriormente, se analizaron variables categóricas como el tipo de contrato, el método de pago y otras características demográficas. Los resultados sugieren que los clientes con contratos mensuales presentan una mayor tendencia a cancelar el servicio, en comparación con aquellos que poseen contratos de mayor duración.

También se analizaron variables numéricas como los cargos mensuales y el tiempo de permanencia del cliente en la empresa (tenure). Las visualizaciones mostraron que los clientes con menor tiempo en la empresa presentan mayores tasas de evasión. Asimismo, los clientes con cargos mensuales más altos parecen tener una mayor probabilidad de cancelar el servicio.

Estas observaciones sugieren que tanto el tipo de contrato como el costo del servicio y la duración de la relación con el cliente pueden influir significativamente en la decisión de cancelar el servicio.

# Conclusiones e Insights

En primer lugar, se observa que los clientes con contratos mensuales tienen una mayor tendencia a cancelar el servicio en comparación con aquellos que poseen contratos de largo plazo. Esto podría deberse a que los contratos mensuales ofrecen mayor flexibilidad para abandonar el servicio.

En segundo lugar, los clientes con menor tiempo de permanencia en la empresa presentan mayores tasas de evasión, lo que sugiere que los primeros meses de relación con el cliente son críticos para su fidelización.

Finalmente, se identificó que los clientes con cargos mensuales más elevados muestran una mayor probabilidad de cancelar el servicio, lo que podría indicar una percepción de alto costo o una falta de valor percibido en el servicio.
